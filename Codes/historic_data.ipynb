```python
top_companies = pd.read_csv("data/top_companies.csv")
```

### sort by market cap to get top companies 

```python
top_companies = top_companies.sort_values(by=["Market Cap"], ascending=False).head(56)
```

### Merge Historic Data with Top Companies

```python
top_companies.rename(columns={"Symbol": "ticker"}, inplace=True)
```

```python
historic_top = pd.merge(top_companies, data, how="left", on=["ticker"])
```

```python
historic_top["date"] = pd.to_datetime(historic_top["date"])
```

### filter by date 

```python
historic_top = historic_top[(historic_top["date"] > "2015")]
```

```python
len(historic_top["Name"].unique())
```

```python
historic_top["Name"].value_counts()
```

#### drop AT&T Inc. 5.350% Global Notes due 2066 because of 202 rows

```python
historic_top = historic_top[
    historic_top["Name"] != "AT&T Inc. 5.350% Global Notes due 2066"
]
```
