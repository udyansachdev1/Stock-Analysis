```python
# installing required packages
import os
import pandas as pd
```

```python
# get list of all files in directory
files = os.listdir("../Data/raw_20/")
files.sort()
files
```

```python
# make sure files doesnt contain System files
files = [file for file in files if file[0] != "."]
files = [file for file in files if file[-4:] == ".csv"]
files
```

```python
# loop through all files and create single dataframe
df_list = []
for file in files:
    temp_df = pd.read_csv("../Data/raw_20/" + file)
    df_list.append(temp_df)
df = pd.concat(df_list)
```

```python
df.sample(10)
```

```python
# reaaraning columns
df = df[
    [
        "Ticker",
        "timestamp",
        "open",
        "high",
        "low",
        "close",
        "volume",
    ]
]
```

```python
df.info()
```

```python
# converting timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df.head(10)
```

```python
# sort by ticker and timestamp
df.sort_values(by=["Ticker", "timestamp"], inplace=True)
```

```python
# save to parquet
df.to_parquet("../Data/processed_20.parquet")
```
